```python

import pandas as pd  
import numpy as np
import urllib.request
import csv
from urllib.parse import urlparse
import urllib.request

```

```python
df = pd.read_csv("C:/Users/PLAYER2/Desktop/ml tweeter/favorite-tweets usable.csv")
```

```python
#finding URL from text/tweets
shorturl = []
pattern = r'(https?:\/\/(?:www\.)?[-a-zA-Z0-9@:%._+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}[-a-zA-Z0-9()@:%_+.~#?&/=]*)' 

df["Text"]=df["Text"].astype(str)

df['links'] = ''
df['links']= df["Text"].str.extract(pattern, expand=True)
for s in df['links']: 
    shorturl.append(s)


shorturl
```

```python
# Finding fill url
df["links"]=df["links"].apply(str)
Flist = []
for m in df["links"]:
    if(m=="nan"):
        Flist.append("Nan")
    else:
        try:
            resp = urllib.request.urlopen(m)
            Flist.append(resp.url)
        except:
            Flist.append("URL ERROR")
```

```python
# full URL list
Flist
```

```python
#finding Domain
dlist =[]
for t in Flist:
    domain = urlparse(t).netloc
    dlist.append(domain)

dlist
```

```python
# Create a Pandas DataFrame.
import pandas as pd
URL_CSV = {
    'Short url':shorturl,
    'Full url' :Flist,
    'Domain list':dlist,
              }
URL = pd.DataFrame(URL_CSV)
```

```python
#converting Data frame to CSV
new=URL.to_csv('URL.csv')
```
